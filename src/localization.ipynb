```python
import numpy as np
import matplotlib.pyplot as plt


def get_loc_mask(r_influ, N):
    """Get a localization mask based on Gaussian decay."""
    tmp = np.zeros((N, N))
    for i in range(1, 3 * r_influ + 1):
        tmp += np.exp(-(i**2) / r_influ**2) * (
            np.diag(np.ones(N - i), i)
            + np.diag(np.ones(N - i), -i)
            + np.diag(np.ones(i), N - i)
            + np.diag(np.ones(i), -(N - i))
        )
    mask = tmp + np.diag(np.ones(N))
    return mask


mask = get_loc_mask(2, 10)
plt.imshow(mask)
plt.colorbar()
```
